```python
def cargar_datos():
    empleados = {}
    with open("empleados.csv", encoding="utf-8") as archivo:
        contenido = archivo.readlines()
        for linea in contenido[1:]:
            if linea.strip():
                dato = linea.strip().split(",")
                empleados[dato[0]] = {
                    "nombre": dato[1],
                    "dias_disponibles": int(dato[2])
                }
    return empleados
```

```python
def formatear_datos(legajo, empleado):
    return f"{legajo},{empleado['nombre']},{empleado['dias_disponibles']}"
```

```python
def mostrar_datos(empleados):
    for legajo, empleado in empleados.items():
        print(formatear_datos(legajo, empleado))
```

```python
def guardar_datos(empleados):
    with open("empleados.csv", "w", encoding="utf-8") as archivo:
        archivo.write("legajo,nombre,dias_disponibles\n")
        for legajo, empleado in empleados.items():
            archivo.write(
                f"{legajo},{empleado['nombre']},{empleado['dias_disponibles']}\n"
            )
```

```python
empleados = cargar_datos()
mostrar_datos(empleados)
```

```python
def solicitar_legajo():
    while True:
        print("\nBOT: Ingrese su número de legajo:")
        legajo = input("USUARIO: ")
        if legajo in empleados:
            return legajo
        print("BOT: El legajo ingresado no existe.")
```

```python
def solicitar_dias():
    while True:
        print("\nBOT: Cuántos días desea solicitar?")
        dato = input("USUARIO: ")
        if dato.isdigit():
            dias = int(dato)
            if dias > 0:
                return dias
            print("BOT: Debe ingresar un número mayor que cero.")
        else:
            print("BOT: Debe ingresar únicamente números.")
```

```python
def procesar_solicitud(legajo, dias_solicitados):
    empleado = empleados[legajo]
    dias_disponibles = empleado["dias_disponibles"]
    print("\nBOT: Consultando saldo de vacaciones...")
    print(
        f"BOT: {empleado['nombre']} posee {dias_disponibles} días disponibles."
    )
    input("Presione enter para continuar")
    if dias_solicitados <= dias_disponibles:
        empleado["dias_disponibles"] -= dias_solicitados
        guardar_datos(empleados)
        print("\nBOT: Solicitud APROBADA.")
        print(
            f"BOT: Le quedan {empleado['dias_disponibles']} días disponibles."
        )
    else:
        print("\nBOT: Solicitud RECHAZADA.")
        print(
            f"BOT: Usted solicitó {dias_solicitados} días y solo dispone de {dias_disponibles}."
        )
```

```python
def continuar():
    while True:
        print("\nBOT: Desea realizar otra solicitud? (SI/NO)")
        respuesta = input("USUARIO: ").upper().strip()
        if respuesta in ("SI", "NO"):
            return respuesta
        print("BOT: Respuesta inválida.")
```

```python
def iniciar_chatbot():
    print()
    print("BOT DE GESTIÓN DE VACACIONES")
    print()
    while True:
        print("\nBOT: Bienvenido al sistema de vacaciones.")
        legajo = solicitar_legajo()
        empleado = empleados[legajo]
        print(f"\nBOT: Hola {empleado['nombre']}." )
        input("Presione enter para continuar")
        dias = solicitar_dias()
        procesar_solicitud(legajo, dias)
        if continuar() == "NO":
            print("\nBOT: Gracias por utilizar el sistema.")
            print("BOT: Hasta luego.")
            break
# Programa principal
iniciar_chatbot()
```
